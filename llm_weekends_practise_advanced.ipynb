```python
import os
from google.colab import userdata
from openai import OpenAI

os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
```

```python
client = OpenAI()
model="gpt-4o-mini"
```

### Summary of Advanced 2026 Logic
The example below demonstrates **Structured Output (JSON)** and **System 2 Thinking (Reasoning)**. In 2026, models are used as "Strategic Analysts" that don't just provide text, but provide machine-readable data that can be used by other software (Agentic Workflows).

### **Explanation of Each Sections**
1. **Strategic Market Analysis (JSON):**

* **The Goal:** Instead of a conversational chat, the model acts as a data engine. It performs "System 2 Thinking" to evaluate if a job (Prompt Engineer) will still exist in 5 years, then extracts skills and formats everything into a JSON object.

* **Agentic Use:** This output is designed so that a Python script could automatically parse the salary_estimate_gbp and plot it on a graph.

2. **Long-Context Legal Reasoning:**

* **The Goal:** This demonstrates the 2026 ability to handle "High-Stakes" documents. It combines Summarization (for SMEs), Closed QA (for specific dates/deadlines), and Classification (to assess regulatory weight).

* **Model Performance:** Even though it is a "mini" model, gpt-4o-mini is highly optimized for these logic-heavy extraction tasks.

### **Estimated Cost & Tokens for Advanced Tasks:**
* Input Tokens: ~250 tokens.
* Output Tokens: ~400 tokens (JSON is often more token-heavy due to formatting).
* Total Cost: Approximately $0.00028 (using 2026 OpenAI Pricing rates).

```python
# =================================================================
# SECTION: ADVANCED 2026 CAPABILITIES (REASONING & STRUCTURED DATA)
# This section demonstrates how modern models handle complex
# career analysis and return data in a structured JSON format.
# =================================================================

# Define the complex multi-step prompt
advanced_prompt = """
Analyze the UK Gen AI job market for a 'Prompt Engineer'.
1. [Reasoning]: Evaluate the long-term viability of this role given model improvements.
2. [Information Extraction]: Identify 3 key skills required.
3. [Output]: Provide a JSON object with 'verdict', 'skills_list', and 'salary_estimate_gbp'.
"""

# API Call using specific pattern (client.responses.create + input)
# Note: For JSON mode, ensure the system prompt explicitly requests JSON.
response_advanced = client.responses.create(
    model=model,
    input=[
        {"role": "system", "content": "You are a strategic career analyst. Always output valid JSON."},
        {"role": "user", "content": advanced_prompt}
    ]
)

# Print Output separately
print("--- Advanced 2026 Capability: Strategic Market Analysis ---")
print(response_advanced.output_text)
```

```python
# =================================================================
# SECTION: MULTI-MODAL CONTEXT REASONING
# This section demonstrates "Long-Context" capabilities where the
# model acts as a legal researcher analyzing a vast dataset.
# =================================================================

legal_prompt = """
[Task: Long-Context Reasoning]
Analyze the provided 2026 UK AI Safety Bill draft (hypothetical context).
1. Summarize the implications for small-to-medium enterprises (SMEs).
2. Identify any 'Closed QA' compliance deadlines mentioned in the text.
3. Classify the regulatory burden as: Low, Moderate, or High.
"""

response_legal = client.responses.create(
    model=model,
    input=[
        {"role": "system", "content": "You are a legal technology consultant."},
        {"role": "user", "content": legal_prompt}
    ]
)

print("\n--- Advanced 2026 Capability: Long-Context Legal Reasoning ---")
print(response_legal.output_text)
```

To explore **2026 Advanced Capabilities**, we move beyond text-only processing into Multimodality (processing images/audio), Autonomous Reasoning (multi-step planning), and Structured Data Synthesis.

Using gpt-4o-mini, you can perform these advanced tasks with the same client.responses.create pattern.

1. **Multimodality: Visual Intelligence**
In 2026, "Information Extraction" isn't limited to text. Models can "see" an image and extract data from it. While the code below uses a URL, you can pass local image data to identify objects or read handwritten notes.

* **Explanation:** This treats an image as a primary input. The model doesn't just describe the image; it performs a Reasoning task based on visual data.

* **Estimated Tokens:** ~1,000 (Images are typically billed as a flat "tile" rate of 85–170 tokens).


```python
# --- Multimodal Vision Task (Fixed Structure) ---
vision_prompt = "Look at this image of a nutrition label. [Extraction]: Extract the total calories and sugar. [Open QA]: Is this a healthy choice for a diabetic person? Explain why."
image_url = "https://upload.wikimedia.org/wikipedia/commons/e/e9/FDA_Nutrition_Facts_Label_2016.png"

try:
    # Note: Use standard 'messages' and type identifiers for gpt-4o-mini
    response_vision = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": vision_prompt},
                    {"type": "image_url", "image_url": {"url": image_url}}
                ]
            }
        ]
    )
    print("--- Advanced Capability 1: Multimodal Vision ---")
    print(response_vision.choices[0].message.content)

except Exception as e:
    print(f"Error encountered: {e}")
```

2. **Agentic Workflow: Multi-Step "Plan & Execute"**
In 2026, models use Chain-of-Thought (CoT). Instead of jumping to an answer, they create a plan. This example asks the model to act as a "Code Auditor."

* **Explanation:** This mimics an "Agent" that thinks through a problem (SQL Injection), fixes it, and then teaches the user.

* **Estimated Tokens:** ~450 (Input: 150, Output: 300).


```python
# --- Agentic Reasoning Task ---
agent_prompt = """
[Task]: Audit this Python function for security flaws:
`def login(user, pw): return db.execute(f"SELECT * FROM users WHERE u='{user}' AND p='{pw}'")`

1. [Reasoning]: Analyze the code for vulnerabilities.
2. [Generation]: Provide a secure, patched version of the code.
3. [Summarization]: Explain the fix in one sentence for a junior developer.
"""

response_agent = client.responses.create(
  model=model,
  input=[
    {"role": "system", "content": "You are a Senior Security Engineer."},
    {"role": "user", "content": agent_prompt},
  ]
)

print("\n--- Advanced Capability: Agentic Security Audit ---")
print(response_agent.output_text)
```

3. **Cross-Language Synthesis**
Advanced 2026 models handle "Code-Switching" and translation with nuance, maintaining technical accuracy across languages.

* **Explanation:** This goes beyond simple translation by applying Classification (identifying what is code vs. what is human language) and Generation in another language.
* **Estimated Tokens:** ~150 (Input: 60, Output: 90).

```python
# --- Cross-Language Synthesis ---
lang_prompt = """
[Task]: Translate this technical requirement into Japanese, but keep the coding variables in English.
Requirement: 'The API must return a 404 error if the user_id is not found in the cache.'
"""

response_lang = client.responses.create(
  model=model,
  input=[
    {"role": "system", "content": "You are a bilingual technical architect."},
    {"role": "user", "content": lang_prompt},
  ]
)

print("\n--- Advanced Capability: Cross-Language Synthesis ---")
print(response_lang.output_text)
```
